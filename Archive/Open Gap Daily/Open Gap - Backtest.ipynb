```python
import pandas as pd
import numpy as np
```

```python
all_data = pd.read_csv('Open-Close Price Long Format.csv')
```

```python
all_data_copy = all_data.pivot_table(index=['Ticker','Date'],columns=['Type'],values='Price').reset_index()
all_data_copy['Prev Close'] = all_data_copy.groupby(['Ticker'])['Close'].shift()
all_data_copy.dropna(inplace=True)
all_data_copy['Open Gap'] = all_data_copy['Open']-all_data_copy['Prev Close']
all_data_copy['Current Day Return'] = all_data_copy['Close']/all_data_copy['Open']-1

all_data_copy['Date'] = all_data_copy['Date'].apply(lambda x: x[:10])
```

```python
all_data_copy.head(10)
```

```python
all_data_copy2 = all_data_copy.pivot_table(index = 'Date',
                                           columns = 'Ticker',
                                           values = ['Open Gap','Current Day Return'])
```

```python
all_data_copy2
```

```python
signal = all_data_copy2['Open Gap'].rolling(101).rank(method='max')>99
signal = pd.DataFrame(signal.stack(),columns=['Signal'])
```

```python
signal = signal.loc[signal['Signal']==True]
```

```python
pnl = pd.DataFrame(all_data_copy2['Current Day Return'].stack(),columns=['Current Day Return']).loc[signal.index]
```

```python
pnl.mean()/pnl.std()
```
